# **Merging 2 dataframes**

```python
employees = pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'Name':['John','Anna','Raj','Linda','Bob'],
    'Department':['Accounting','IT','IT','HR','Accounting']
})

Salaries = pd.DataFrame({
    'employee_id': [1,2,3,6,7],
    'salaries' : [60000,80000,65000,70000,90000],
    'bonus' : [5000,10000,7000,8000,12000]
})
employees
```

```python
Salaries
```

```python
pd.merge(employees, Salaries)
```

```python
pd.merge(employees, Salaries,on = "employee_id",how= 'inner')
```

```python
pd.merge(employees, Salaries,on = "employee_id",how= 'outer')
```

```python
# based on employee
pd.merge(employees, Salaries,on = "employee_id",how= 'left')
```

```python
# based on Salary
pd.merge(employees, Salaries,on = "employee_id",how= 'right')
```

# **GroupBy** - on categorical data
> it creates an object that we can't see

```python
data = {
    'Category' : ['A','B','A','B','A','B','A','B'],
    'Store' : ['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales' : [100,200,150,300,250,350,200,300],
    'Quantity' : [10,15,12,18,8,20,15,15],
    'Date' : pd.date_range('2023-01-01',periods=8)}
df = pd.DataFrame(data)
df
```

> Q- Group by Category and Calculate  the sum of Sales

```python
df.groupby('Category')['Sales'].sum()
```

```python
cat = df.groupby('Category')
cat.get_group('A')
```

```python
cat = df.groupby('Category')
cat
for i,v in cat:
  print(i)
  print(v)
```

> Q - Group by Store and calculate the sum of sales

```python
df.groupby('Store')['Sales'].sum()
```

> Q- Group multiple columns

> Q- group by Category and store

```python
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```

# **Pivot Tables**
> we use ***agg func***

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'dates': pd.date_range('2023-01-01', periods=20),
    'product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam'] * 4,
    'region': ['North', 'South', 'East','West'] * 5,
    'sales': np.random.randint(100, 1000, 20),
    'units': np.random.randint(1, 10, 20),
    'rep': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'] * 4
}
df_new = pd.DataFrame(data)

df_new['Month'] = df_new['dates'].dt.month_name()
df_new['Quarter'] = 'Q' + df_new['dates'].dt.quarter.astype(str)
df_new
```

```python
pd.pivot_table(df_new,values ='sales',index = 'region',columns = 'product')
```

```python
pd.pivot_table(df_new,values =['sales','units'],index = 'region',columns = 'product')
```

# **Cross Tabs**
> we use ***counting func***

```python
pd.crosstab(df_new['region'],df_new['product'])
```

````python
import numpy as np
# col:[Age , math marks , Science marks]
data = np.array([
    [18,85,78],
    [19,92,88],
    [17,76,95],
    [18,65,70],
    [20,90,85]
])
````

````python
data.shape
````

````python
# avg age of student
np.mean(data[:,0])
````

````python
# extract math marks of all student
data[:,1]
````

````python
# find the highest science marks
np.max(data[:,2])
````

````python
# get the details of student who scored highest in math
data[data[:,1]> 90]
````

````python
# Increase math marks of all student by 5
data[:,1] += 5
data
````

````python
# find how many student are younger than 19
len(data[data[:,0]<19])
````

````python
#calculate the average marks in each subject(column-wise mean)
np.mean(data[:,1:],axis=0)
````

````python
#get data of student who scored at least 80 in both subjects
data[(data[:,1]>=80) & (data[:,2]>=80)]
````

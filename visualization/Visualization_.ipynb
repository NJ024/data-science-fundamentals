# **Object oriented way**

### Matplotlib allows the aspect ratio ,DPI and figure size to be specified when the figure object is created. <br>
### You can use <b>figsize and dpi keyword. </b>arguments <br>

### figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel-per-inch).

````python
# only canvas is selected -> no axis
fig = plt.figure(figsize=(12,8),dpi=500)
axis1 =fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 =fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
````

# **Distribution Plot**

### we use distribution plot to visualize the distribution of <b>quantitative data

<br>

These plots are:
*   distplot/histplot
*   jointplot
*   pairplot
*   rugplot



<h2><b>Seaborn<b></h2>

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = sns.load_dataset('tips')
df
````

### **displot**

````python
df['size'].unique()
````

````python
sns.displot(df['total_bill'],bins = 20)
````

### **histplot**

### **kde--> kernal density estimator** --> tells skewness / distribution

````python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde= True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde= True)
````

### **jointplot**

````python
sns.jointplot(x = 'total_bill', y= 'tip',data= df,kind = "scatter")
````

````python
sns.jointplot(x = 'total_bill', y= 'tip',data= df,kind = "kde")
````

````python
sns.jointplot(x = 'total_bill', y= 'tip',data= df,kind = "hex")
````

````python
sns.jointplot(x = 'total_bill', y= 'tip',data= df,kind = 'reg',color = 'black')
````

### **pairplot**

````python
sns.pairplot(df)
````

### **hex--> category wise divide / differentiate**

````python
sns.pairplot(df,hue = 'sex',palette = 'rainbow')
````

### **rugplot**

````python
sns.rugplot(df['total_bill'])
````

# **Categorical Data Plots** using seaborn

```

*   violinplot

*   countplot

```






````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = sns.load_dataset("tips")
df.head()
````

### **countplot**

````python
sns.countplot(x=df['sex'],hue = df['smoker'])
````

### **barplot** --> x= categorical and y = quantitative

````python
sns.barplot(x = df['sex'], y = df['total_bill'])
````

````python
import numpy as np
sns.barplot(x = df['sex'], y = df['tip'], estimator= np.sum)
````

### **boxplot & violinplot**

are used to show the **distribution of categorical data**. <br>
 A box plot( box and whisker plot) shows** distribution of quantitative data** in a way that facilitates** comparisons between variables or across levels of categorical variable**.<br>
*The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution , except for the points that are determined to be 'outliers' using a method that function of the inter- quartile range.*

````python
sns.boxplot(x= df['tip'],y = df['day'], palette= 'rainbow')
````

````python
sns.violinplot(x = df['day'], y = df['total_bill'])
````

### **stripplot and swarmplot**

The strip plot will draw a scatterplot where 1 variable is categorical.<br>
A strip plot can be drawn on its own , but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.
<br>
<br>
The swarmplot is similar to stripplot(), but the points are adjusted (only the categorical axis) so that they don't  overlap . This gives a better representation of the distribution of values, althrough it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them)

````python
sns.stripplot(x = df['day'], y = df['total_bill'])
````

````python
sns.swarmplot(x = df['day'], y = df['total_bill'])
````

````python
sns.violinplot(x='tip', y ='day',data = df)
sns.swarmplot(x='tip', y='day',data = df,color = 'yellow ')
````
